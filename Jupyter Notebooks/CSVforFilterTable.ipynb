```python
#Import Dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine_df = pd.read_csv("../Resources/winequality.csv")

wine_df.head()
```

```python
#get out columns we want in the order we want

wine_df = wine_df[["color", "quality", "volatile acidity", "density", "alcohol", "pH", "residual sugar", 
                   "sulphates", "chlorides"]]

wine_df
```

```python
#Rename columns

wine_df = wine_df.rename(columns={"color":"Color", "quality":"Quality", "volatile acidity":"Volatile Acidity", 
                                  "density":"Density", "alcohol":"Alcohol", "pH":"pH", "residual sugar":"Residual Sugar",
                                  "sulphates":"Sulphates", "chlorides":"Chlorides"})

wine_df
```

```python
#Sort Descending by Quality

wine_df = wine_df.sort_values(["Quality","Alcohol"], ascending=False)

wine_df.head(20)
```

```python
#Round to 2 decimal places
wine_df = wine_df.round(2)

wine_df
```

```python
#Export to csv without the pandas index, but with the header

wine_df.to_csv("../Resources/winequality1.csv", index=False, header=True)
```
